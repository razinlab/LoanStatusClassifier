#Model Evaluation

Now that our selection and tuning is done, let's evalute final performance of the model on the test set:

```python
x_test = test_set.drop('loan_status', axis=1)
y_test = test_set['loan_status'].copy()

y_hat = final_model.predict(x_test)

cm = confusion_matrix(y_test, y_hat)
accuracy = accuracy_score(y_test, y_hat)

print(f'XGB Classifier Report: \n{classification_report(y_test, y_hat)}')
sns.heatmap(cm, annot=True, fmt='d', cmap='flare',
          xticklabels=['Rejected', 'Approved'],
          yticklabels=['Rejected', 'Approved'])
plt.title(f'Confusion Matrix for XGB Classifier')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

print(f'XGB Classifier Accuracy: {accuracy:.3f}')
```

Works great, however model improvement from parameter tuning was minimal.
